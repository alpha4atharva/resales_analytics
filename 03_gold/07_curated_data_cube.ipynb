```python
spark.sql("""
CREATE OR REPLACE TABLE de_mini_project.gold.sales_data_cube AS
SELECT 
    s.transaction_date,
    s.transaction_id,
    s.product_id,
    p.item_name,
    p.category,
    p.cost_price,
    p.profit_per_unit,
    s.store_id,
    st.location AS store_location,
    st.city,
    st.region,
    s.customer_id,
    c.name,
    s.sold_qty,
    s.actual_sold_price,
    s.total_revenue,
    s.is_returned    
FROM de_mini_project.gold.fact_sales s
INNER JOIN de_mini_project.gold.dim_product p ON s.product_id = p.product_id
INNER JOIN de_mini_project.gold.dim_store st ON s.store_id = st.store_id
LEFT JOIN de_mini_project.gold.dim_customer c ON s.customer_id = c.customer_id
""")
```

```python
result = spark.sql("""
    SELECT *
    FROM de_mini_project.gold.sales_data_cube
""")
display(result)
```

```python
spark.sql("""
CREATE OR REPLACE TABLE de_mini_project.gold.inventory_data_cube AS
SELECT 
    inv.last_audit_date,
    inv.last_sale_date,
    inv.product_id,
    p.item_name,
    p.category,
    p.cost_price,
    p.profit_per_unit,
    inv.store_id,
    s.location AS store_location,
    s.city,
    s.region,
    inv.stocks_qty,
    inv.is_out_of_stock,
    inv.is_slow_moving   
FROM de_mini_project.gold.fact_inventory inv
INNER JOIN de_mini_project.gold.dim_product p ON inv.product_id = p.product_id
INNER JOIN de_mini_project.gold.dim_store s ON inv.store_id = s.store_id
""")
```

```python
result = spark.sql("""
    SELECT *
    FROM de_mini_project.gold.inventory_data_cube
""")
display(result)
```
