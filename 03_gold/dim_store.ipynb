```python
# 1. Ensure the Gold database exists
spark.sql("CREATE DATABASE IF NOT EXISTS de_mini_project.gold")

# 2. Create the Gold Store Dimension
# We drop 'contact' to keep the table clean for the Sales Dashboard
spark.sql("""
CREATE OR REPLACE TABLE de_mini_project.gold.dim_store AS
SELECT 
    store_id,
    location,
    city,
    region
FROM de_mini_project.silver.store
""")

# 3. Validation: Check the geographic hierarchy
display(spark.sql("SELECT region, city, COUNT(store_id) as store_count FROM de_mini_project.gold.dim_store GROUP BY 1, 2"))
```
