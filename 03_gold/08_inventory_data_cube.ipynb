```python
# --- CORRECTED: INVENTORY DATA CUBE ---
# Aggregates strictly on the Inventory grain. 

spark.sql("""
CREATE OR REPLACE TABLE de_mini_project.gold.inventory_cube AS
SELECT 
    i.product_id,
    i.store_id,
    
    -- KPI 6 & 10: Inventory Risks
    MAX(i.stocks_qty) AS current_stock,
    MAX(i.is_out_of_stock) AS oos_flag,
    MAX(i.is_slow_moving) AS slow_moving_flag

FROM de_mini_project.gold.fact_inventory i
GROUP BY 1, 2
""")
```
