```python
# Setup: Ensure the Gold database exists
spark.sql("CREATE DATABASE IF NOT EXISTS de_mini_project.gold")

# 1. Drop existing table to avoid duplicates or schema mismatches
spark.sql("DROP TABLE IF EXISTS de_mini_project.gold.dim_product")

# 2. Create the Gold Dimension Table
# This fulfills KPI #2: Profit Per Product (Selling Price - Cost Price)
spark.sql("""
CREATE OR REPLACE TABLE de_mini_project.gold.dim_product AS
SELECT 
    product_id,
    item_name,
    category,
    -- We keep these for KPI #2 (Profit)
    base_price AS cost_price,
    marked_price AS selling_price,
    (marked_price - base_price) AS profit_per_unit
FROM de_mini_project.silver.product
""")

# 3. Validation Check
display(spark.sql("SELECT * FROM de_mini_project.gold.dim_product LIMIT 5"))
```
