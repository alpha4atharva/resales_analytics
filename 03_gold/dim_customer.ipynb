```python
# 1. Drop existing table to avoid duplicates
spark.sql("DROP TABLE IF EXISTS de_mini_project.gold.dim_customer")

# 2. Create the Gold Customer Dimension Table
# We use UNION ALL to add the 'Guest' placeholder row
spark.sql("""
CREATE OR REPLACE TABLE de_mini_project.gold.dim_customer AS
SELECT 
    customer_id,
    name,
    gender
FROM de_mini_project.silver.customer

UNION ALL

SELECT 'UNKNOWN_CUSTOMER', 'Guest Customer', 'Unknown'
""")

# 3. Validation: Verify the 'Unknown' row exists
display(spark.sql("SELECT * FROM de_mini_project.gold.dim_customer WHERE customer_id = 'UNKNOWN_CUSTOMER'"))
```
