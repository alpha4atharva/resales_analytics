```python
# KPI 1: TOTAL SALES
# Definition: Total money collected from customers for all products sold.
kpi_1 = spark.sql("""
    SELECT 
        ROUND(SUM(total_revenue), 2) AS Total_Sales_INR 
    FROM de_mini_project.gold.fact_sales
""")
display(kpi_1)

# KPI 2: PROFIT PER PRODUCT
# Definition: Selling price minus the cost of the item.
kpi_2 = spark.sql("""
    SELECT 
        item_name, 
        category, 
        profit_per_unit AS Unit_Profit_INR 
    FROM de_mini_project.gold.dim_product 
    ORDER BY profit_per_unit DESC
""")
display(kpi_2)

# KPI 3: TOP SELLING CATEGORY
# Definition: Which group (e.g. Electronics) brings the most money?
kpi_3 = spark.sql("""
    SELECT 
        p.category, 
        ROUND(SUM(s.total_revenue), 2) AS Total_Category_Revenue 
    FROM de_mini_project.gold.fact_sales s
    JOIN de_mini_project.gold.dim_product p ON s.product_id = p.product_id
    GROUP BY 1 
    ORDER BY 2 DESC 
    LIMIT 1
""")
display(kpi_3)

# KPI 4: BASKET SIZE
# Definition: Average number of items bought per single transaction.
kpi_4 = spark.sql("""
    SELECT 
        ROUND(AVG(sold_qty), 2) AS Average_Items_Per_Basket 
    FROM de_mini_project.gold.fact_sales
""")
display(kpi_4)

# KPI 5: RETURN RATE
# Definition: Percentage of total sales that were returned by customers.
kpi_5 = spark.sql("""
    SELECT 
        ROUND((SUM(is_returned) / COUNT(*)) * 100, 2) AS Return_Rate_Percentage 
    FROM de_mini_project.gold.fact_sales
""")
display(kpi_5)

# KPI 6: OUT-OF-STOCK (COUNT)
# Definition: Count of unique products with zero units in stock.
kpi_6 = spark.sql("""
    SELECT 
        COUNT(DISTINCT product_id) AS Out_of_Stock_Product_Count 
    FROM de_mini_project.gold.fact_inventory 
    WHERE is_out_of_stock = 1
""")
display(kpi_6)

# KPI 7: STORE PERFORMANCE
# Definition: Ranking of the 5 stores based on daily sales volume.
kpi_7 = spark.sql("""
    SELECT 
        st.location, 
        st.region, 
        ROUND(SUM(sa.total_revenue), 2) AS Total_Sales_Volume 
    FROM de_mini_project.gold.fact_sales sa
    JOIN de_mini_project.gold.dim_store st ON sa.store_id = st.store_id
    GROUP BY 1, 2 
    ORDER BY 3 DESC
""")
display(kpi_7)

# KPI 8: DISCOUNT IMPACT
# Definition: Money lost due to differences in marked vs. sold price.
kpi_8 = spark.sql("""
    SELECT 
        ROUND(SUM(discount_loss_amount), 2) AS Total_Discount_Loss_INR 
    FROM de_mini_project.gold.fact_sales
""")
display(kpi_8)

# KPI 9: REPEAT CUSTOMER COUNT
# Definition: Number of unique users with more than one purchase.
kpi_9 = spark.sql("""
    SELECT 
        COUNT(*) AS Repeat_Customer_Count
    FROM (
        SELECT customer_id 
        FROM de_mini_project.gold.fact_sales 
        GROUP BY customer_id 
        HAVING COUNT(DISTINCT transaction_id) > 1
    )
""")
display(kpi_9)

# KPI 10: SLOW-MOVING INVENTORY
# Definition: Products with zero sales recorded in the last 30 days.
kpi_10 = spark.sql("""
    SELECT 
        COUNT(DISTINCT product_id) AS Slow_Moving_Product_Count 
    FROM de_mini_project.gold.fact_inventory 
    WHERE is_slow_moving = 1
""")
display(kpi_10)
```
