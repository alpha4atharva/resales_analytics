```python
spark.sql("DROP TABLE IF EXISTS de_mini_project.raw.store")
# Define new table with custom columns
spark.sql("""
CREATE TABLE IF NOT EXISTS de_mini_project.raw.store (
    store_id STRING,
    location STRING,
    city STRING,
    region STRING,
    contact STRING
)
""")

# Insert selected columns from source table
spark.sql("""
INSERT INTO de_mini_project.raw.store (store_id, location, city, region, contact)
SELECT 
    st_id_ AS store_id,
    regexp_replace(location_name_address, '^TechNova - ', '') AS location,
    split(city_region, '_')[0] AS city,
    split(city_region, '_')[1] AS region,
    mgr_contact_ AS contact
FROM de_mini_project.azure_blob_storage.stores
""")
```
