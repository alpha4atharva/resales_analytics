```python
spark.sql("DROP TABLE IF EXISTS de_mini_project.raw.inventory")
# Define new table with custom columns
spark.sql("""
CREATE TABLE IF NOT EXISTS de_mini_project.raw.inventory (
    sku_id STRING,
    st_id STRING,
    stocks_qty INTEGER,
    last_audit_date DATE   
)
""")

# Insert selected columns from source table
spark.sql("""
INSERT INTO de_mini_project.raw.inventory (sku_id, st_id, stocks_qty, last_audit_date)
SELECT 
    _sku_id_ AS sku_id,
    st_id_ref AS st_id,
    _stock_on_hand_ AS stocks_qty,
    try_to_date(last_audit_dt, 'yyyyMMdd') AS last_audit_date
FROM de_mini_project.azure_blob_storage.inventory
""")
```
