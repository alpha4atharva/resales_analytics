```python
# Define new schema
spark.sql("CREATE SCHEMA IF NOT EXISTS de_mini_project.raw")
```

```python
spark.sql("DROP TABLE IF EXISTS de_mini_project.raw.customer")
# Define new table with custom columns
spark.sql("""
CREATE TABLE IF NOT EXISTS de_mini_project.raw.customer (
    customer_id STRING,
    name STRING,
    contact_info STRING,
    joined_date DATE,
    gender STRING    
)
""")

# Insert selected columns from source table
spark.sql("""
INSERT INTO de_mini_project.raw.customer (customer_id, name, contact_info, joined_date, gender)
SELECT 
    _cust_ref_id_ AS customer_id,
    full_name AS name,
    _contact_info_ AS contact_info,
    COALESCE(
        try_to_date(joined_date_, 'dd-MM-yyyy'),
        try_to_date(joined_date_, 'MMM dd, yyyy'),
        try_to_date(joined_date_, 'yyyy.MM.dd')
    ) AS joined_date,
    CASE WHEN gender_code = 'M' THEN 'Male'
         WHEN gender_code = 'F' THEN 'Female'
         ELSE 'Unknown'
    END AS gender
FROM de_mini_project.azure_blob_storage.customers
""")
```
