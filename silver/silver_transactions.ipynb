```python
spark.sql("DROP TABLE IF EXISTS de_mini_project.raw.returns")
# Define new table with custom columns
spark.sql("""
CREATE TABLE IF NOT EXISTS de_mini_project.raw.returns (
    return_id STRING,
    transaction_id STRING,
    date DATE,
    reason STRING
)
""")

# Insert selected columns from source table
spark.sql("""
INSERT INTO de_mini_project.raw.returns (return_id, transaction_id, date, reason)
SELECT 
    rtn_id_no AS return_id,
    orgnl_trxn_id_ AS transaction_id,
    COALESCE(
        try_to_date(rtn_date_string, 'dd-MM-yyyy'),
        try_to_date(rtn_date_string, 'yyyy.MM.dd'),
        try_to_date(rtn_date_string, 'dd-MMM-yy'),
        try_to_date(rtn_date_string, 'MM/dd/yyyy'),
        try_to_date(rtn_date_string, 'MMM dd, yyyy')
    ) AS date,
    rsn_code_ AS reason
FROM de_mini_project.azure_blob_storage.transactions
""")
```
